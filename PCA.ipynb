```python
import numpy as np
import csv
import pandas as pd

######### data reading #############

train_x_data = pd.read_csv('train_images.csv', header=None)
train_x_data = train_x_data.to_numpy()
print('train_x_data',train_x_data)
```

```python
mean_of_features = np.mean(train_x_data,axis = 0) #mean from each column/feature
#calculate variance matrix 
diff_from_mean = train_x_data - mean_of_features # get (x-mean)
# print(diff_from_mean)
covariance_matrix = np.cov(diff_from_mean.T) # get covariance x-mean*x-mean
# print(covariance_matrix)
eigen_values,eigen_vectors = np.linalg.eig(covariance_matrix) #get eigen values and vectors
#sorting eigen values in descending order
sorted_indices_eigen_values = np.argsort(eigen_values) #get indices in descending order of eigen values
sorted_eigen_values = eigen_values[sorted_indices_eigen_values[::-1]] #extract the eigen values
sorted_eigen_vectors = eigen_vectors[sorted_indices_eigen_values[::-1]] #extract the eigen vectors
print(sorted_eigen_values)
```

```python
sorted_eigen_vectors1 = sorted_eigen_vectors [:,:1] #get first 1 component #you can change component here
print('sorted_eigen_vectors1:',sorted_eigen_vectors1.shape)
print('diff_from_mean:',diff_from_mean.shape)
```

```python
new_train_x_data = np.dot(sorted_eigen_vectors1.T,diff_from_mean.T).T #take dot product of eigen vector and diff_from_mean vector
print(new_train_x_data.shape)
```

```python
new_train_x_data_real = np.real(new_train_x_data) #getting only the real numbers and discarding imaginary numbers
```

```python
#writing the newly reduced dimension data to file
df = pd.DataFrame(new_train_x_data_real)
df.to_csv('1_dim_train_data.csv',header=False,index=False)
```

```python
############ test data #####################
test_x_data = pd.read_csv('test_images.csv', header=None)
test_x_data = test_x_data.to_numpy()
# print('test_x_data',train_x_data)
```

```python
diff_from_mean_test = test_x_data - mean_of_features # use the mean vector from PCA and subtract from test x
new_test_x_data = np.dot(sorted_eigen_vectors1.T,diff_from_mean_test.T).T ##take dot product of eigen vector and diff_from_mean vector
new_test_x_data_real = np.real(new_test_x_data) #getting only the real numbers and discarding imaginary numbers
```

```python
df2 = pd.DataFrame(new_test_x_data_real)
df2.to_csv('1_dim_test_data.csv',header=False,index=False)
```
